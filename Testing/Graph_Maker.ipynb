# Notebook for making testing graphs

```python
import matplotlib.pyplot as plt
import cv2

Time_execution = [33.4,43,43,31,5]
```

```python
def Graph(Time_Execution,tries,CHATGPT,models):
    fig = plt.figure(figsize = (25,5))

    ax1 = fig.add_subplot(1,2,1)

    aux_model = str(models)
    plt.plot(Time_execution,color='blue', linestyle='dashed', linewidth = 3, marker='o', markerfacecolor='red', markersize=12)
    plt.title("The number of models used is " +  aux_model,loc = "right")
    plt.title(CHATGPT,loc = "left")
    plt.xticks([])
    plt.ylabel("Seconds")
    
    plt.legend(["Script"])

    
    print("The average amount of tries is:",  tries/len(Time_Execution))
    print("\n")
    plt.grid()
    plt.savefig("Graphs/" + CHATGPT + "_" + "NModels_" + aux_model + ".jpg")
    
    plt.show()
```

```python
Graph(Time_execution,5,"ChatGPT3.5 Turbo",5)
```
